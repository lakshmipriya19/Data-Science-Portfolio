```python
# Import libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, max as spark_max

# Start Spark session
spark = SparkSession.builder.appName("Mini_Assignment_2").getOrCreate()

# Load dataset
df = spark.read.csv("Flight Dataset - CSV(in).csv", header=True, inferSchema=True)

# Show data
df.show(5)
```

Task 1: Create a function that gives back how many flights arrived earlier than expected. 

```python
def early_arrivals(df):
    return df.filter(col("ARR_DELAY") < 0).count()

print("Task 1 Output:", early_arrivals(df))
```

Task 2: Create a function that determines the typical departure time for flights over 2000 miles. 

```python
def typical_departure_time_mode(df):
    return df.filter(col("DISTANCE") > 2000) \
             .groupBy("DEP_TIME") \
             .count() \
             .orderBy("count", ascending=False) \
             .first()[0]

print("Task 2:", typical_departure_time_mode(df))
```

Task 3: Create a function that gives back the proportion of flights that have arrival delays longer than 60 minutes. 

```python
def delayed_flight_proportion(df):
    total = df.count()
    delayed = df.filter(col("ARR_DELAY") > 60).count()
    return delayed / total

print("Task 3 Output:", round(delayed_flight_proportion(df),4))
```

Task 4: Create a function that gives the average airtime for flights that left earlier than 9:00 am. 

```python
def avg_airtime_before_9(df):
    result = df.filter(col("DEP_TIME") < 9.0).select(avg("AIR_TIME"))
    return result.collect()[0][0]

print("Task 4 Output:", round(avg_airtime_before_9(df),2))
```

Task 5: Create a function that determines the maximum arrival delay for flights that did not experience a delay upon departure

```python
def max_arrival_delay_no_dep_delay(df):
    result = df.filter(col("DEP_DELAY") <= 0).select(spark_max("ARR_DELAY"))
    return result.collect()[0][0]

print("Task 5 Output:", max_arrival_delay_no_dep_delay(df))
```

Assumptions made:

1. Negative ARR_DELAY indicates early arrival.
2. Typical departure time interpreted as most frequent departure time (mode).
3. DEP_DELAY <= 0 considered no departure delay.
4. DEP_TIME values are represented in decimal hours.
5. Missing/null values ignored automatically.
